```python
def replace_in_file(file_path, str1, str2):
    #read input file
    f_in = open(file_path, "rt")
    #read file contents to string
    data = f_in.read()
    #replace all occurrences of the required string
    data = data.replace(str1, str2)
    #close the input file
    f_in.close()
    #open the same file in write mode
    f_out = open(file_path, "wt")
    #overrite the file with the resulting data
    f_out.write(data)
    #close the file
    f_out.close()
```

```python
replace_in_file( '.\data_simulation\S020\experiment_n_03\S020_03_table_49.txt', '---', ',')
```

```python
data_sim = root / 'data_simulation'

#sim_path = data_sim / 'S020'

table_files = data_sim.glob('**/*table*.txt')
for i in table_files:
    replace_in_file(i, 'rho', 'col_')
```

```python
data_sim = root / 'data_simulation'

#sim_path = data_sim / 'S020'

values_files = data_sim.glob('**/*value*.txt')
df = pd.DataFrame()
for i in values_files:
    inf_list = i.stem.split('_')
    inf_list.pop(2)
    #print(inf_list)
    df_aux = pd.read_csv(i, sep=' ', header=None)
    df_aux.columns = ['ind', '_'.join(inf_list) ]
    df_aux.index = df_aux['ind']
    df_aux.drop(columns='ind', inplace = True)
    df = pd.concat([df, df_aux], axis =1)
    #print(df.head())

```
